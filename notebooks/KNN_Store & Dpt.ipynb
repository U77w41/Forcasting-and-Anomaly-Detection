```python
%%time
path = "store-department-data"
cols_to_norm = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI', 'Unemployment']
stores_depts = []
print("--------------- Starting Execution -------------")
for i in range(1, 46):
    dirc = f"{path}/store_{i}"
    stre = {}
    for j in range(1, 100):
        try:
            name =  f"store_{i}_dept_{j}"
            st_dt = pd.read_csv(f"{dirc}/Store_{i}_dept_{j}.csv")
            st_dt = st_dt.set_index("Date")
            #st_dt[cols_to_norm] = st_dt[cols_to_norm].apply(lambda x: (x - x.min())/(x.max()-x.min())
            stre[f"dept_{j}"] = st_dt
            globals()[name] = st_dt
            print(f"Read: Store_{i}_dept_{j}.csv")
        except:
            print(f"Store_{i}_dept_{j} skipped.---------------------")
    stores_depts.append(stre)
    print(f"########## Collected for Store {i} ##########")
print("---------------- End of Execution -------------")       
```

```python
metrics_df = stores_depts[10]["dept_1"]
metrics_df.columns
```

```python
from sklearn.preprocessing import MinMaxScaler
mms = MinMaxScaler()
```

```python
to_model_columns=metrics_df.drop(["Store", "Dept"], axis = 1)
to_model_columns[['Weekly_Sales', 'Temperature','Fuel_Price', 'CPI', 'Unemployment']] = mms.fit_transform(to_model_columns[['Weekly_Sales', 'Temperature','Fuel_Price', 'CPI', 'Unemployment']])
```

```python
to_model_columns
```

```python
X = to_model_columns.values
```

```python
X
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
# instantiate model
nbrs = NearestNeighbors(n_neighbors = 3)
# fit model
nbrs.fit(X)
```

```python
# distances and indexes of k-neaighbors from model outputs
distances, indexes = nbrs.kneighbors(X)
# plot mean of k-distances of each observation
plt.figure(figsize=(20,8))
plt.plot(distances.mean(axis =1))
```

```python
# visually determine cutoff values > 0.3
outlier_index = np.where(distances.mean(axis = 1) > 0.3)
#outlier_index
```

```python
# filter outlier values
outlier_values = metrics_df.iloc[outlier_index]
outlier_values
```

```python
%%time
distance = []
outliers = []
percentage = []
print("--------------- Starting Execution -------------")
for i in range(1, 46):
    stor = stores_depts[i-1]
    dis = {}
    out = {}
    perc = {}
    for j in range(1, 100):
        try:
            X = stor[f"dept_{j}"]
            X = X.drop(["Store", "Dept"], axis = 1)
            nbrs = NearestNeighbors(n_neighbors = 3)
            knn_model = nbrs.fit(X)
            distances, indexes = nbrs.kneighbors(X)
            dis[f"dept_{j}"] = distances.mean(axis = 1)
            out[f"dept_{j}"] = X[distances.mean(axis = 1) >= 200]
            perc[f"dept_{j}"] = sum(distances.mean(axis = 1) >= 200)/X.shape[0]*100
            print(f'No. of outliers in store {i} Dept {j} is {sum(distances.mean(axis = 1) >= 200)/X.shape[0]*100}%')
        except:
            print(f"Store_{i}_dept_{j} data not avialable.---------------------")
    distance.append(dis)
    outliers.append(out)
    percentage.append(perc)
print("---------------- End of Execution -------------")  
```
